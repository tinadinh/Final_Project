```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps

# Import API key
from config_PHU import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Store the CSV you saved created in part one into a DataFrame.
phu_df = pd.read_csv("/Users/michellegoldfinger/Desktop/Data_Analysis_and_Visualization/Module_20/Final_Project/Week_2/Datasource/Tables/Covid19_Datasources/phu_joined.csv")
phu_df.head()
```

```python
# Drop unrelated columns from dataframe

columns_to_drop=['id','reporting_phu_address', 'outbreak','week', 
                 'month','year', 'age_group', 'age_group_count','gender','gender_count']
                
               
cleaned_phu_df = phu_df.drop(columns= columns_to_drop)
cleaned_phu_df.head()
```

```python
# Group by phu_id and count records
phu_counts  = cleaned_phu_df.groupby(["phu_id", "reporting_phu"]).count()["outcome"]
phu_counts
```

```python
# phu_counts series to dataframe
phu_counts_df = phu_counts.to_frame()
phu_counts_df
```

```python
# Merge Dataframes
merged_phu_df = pd.merge(cleaned_phu_df, phu_counts_df, on=["phu_id", "phu_id"])
merged_phu_df.tail()
```

```python
# Heatmap of PHUs
# Get the latitude and longitude.
locations = merged_phu_df[["reporting_phu_latitude", 
                            "reporting_phu_longitude"]]
                           
# Get outcomes.
outcome = merged_phu_df["outcome_y"]

# Assign the figure variable.
fig = gmaps.figure()
                           
# Assign the heatmap variable.
heat_layer = gmaps.heatmap_layer(locations, weights=outcome,point_radius=20.0)

# Add the heatmap layer.
fig.add_layer(heat_layer)
# Call the figure to plot the data.
fig
```
